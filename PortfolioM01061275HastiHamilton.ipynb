###  Savings Projection (3 Years)

Compute final savings after 3 years given:
- Save £10 every week for 3 years
- Year 1 bonus: +£20 each week
- Every 2 weeks: −£0.50 (sister)
Assume 52 weeks per year; use arithmetic and augmented assignment; no conditionals.

**Example Output:**
```
Final savings after 3 years: £2561.0
```

````python
deduction_weeks=(52//2)  

regular_weeks = 52-deduction_weeks 

year1_regular_total = 30*(regular_weeks)

years2_3_regular_total = 10*(2*(regular_weeks))

years2_3_deduction_total = 9.5*(2*(deduction_weeks))

year1_deduction_total = 29.5*(deduction_weeks)

final_savings = year1_regular_total + years2_3_regular_total + years2_3_deduction_total + year1_deduction_total
````

````python
print(f'Final savings after 3 years: £{final_savings}')
````

## Explanation of the savings Projection

In order to solve this problem, I used: variables; integer and float data types; arithmetic operations ; augmented assignment ; parentheses to control order of operations; constants (e.g., 52 weeks per year); and formatted output using print() with f-strings.


 `deduction_weeks = (52 // 2)`  
This line uses integer division (`//`) to find how many weeks per year include a sister deduction.  
Since a deduction happens every 2 weeks, 52 divided by 2 equals **26** deduction weeks.

 `regular_weeks = 52 - deduction_weeks`  
This calculates the number of weeks without any sister deduction.  
There are 52 weeks in total, minus 26 deduction weeks, leaving **26 regular weeks**.

 `year1_regular_total = 30 * (regular_weeks)`  
This finds the total savings in **Year 1** during the regular weeks.  
Each regular week in Year 1 earns £10 + £20 bonus = **£30**, so £30 × 26 = **£780**.

 `years2_3_regular_total = 10 * (2 * (regular_weeks))`  
This calculates the total savings in **Years 2 and 3** during regular weeks.  
There is no bonus in these years, so it’s £10 per week.  
Two years × 26 regular weeks = 52 weeks, and £10 × 52 = **£520**.

 `years2_3_deduction_total = 9.5 * (2 * (deduction_weeks))`  
This calculates the total savings for **Years 2 and 3** during deduction weeks.  
Each deduction week has £10 − £0.50 = **£9.50** saved.  
Two years × 26 deduction weeks = 52 weeks, and £9.5 × 52 = **£494**.

`year1_deduction_total = 29.5 * (deduction_weeks)`  
This calculates the total savings in **Year 1** during deduction weeks.  
Each deduction week has £30 − £0.50 = **£29.50** saved, so £29.5 × 26 = **£767**.

 `final_savings = year1_regular_total + years2_3_regular_total + years2_3_deduction_total + year1_deduction_total`  
This adds up all the amounts from the previous calculations to find the total savings after 3 years.  
£780 + £520 + £494 + £767 = **£2561.0** total savings.

 `print(f'Final savings after 3 years: £{final_savings}')`  
This prints the result using an **f-string**, which allows variables to be displayed inside text.  
The output will show:  
**Final savings after 3 years: £2561.0**

### Payroll Statement

Ask for: employee name, hours worked, hourly rate, federal tax rate, state tax rate.
Print a formatted payroll slip with gross pay, each deduction, and net pay (2 d.p.).

**Example Output:**
```
Employee: Smith
Gross pay: £97.50
Federal withholding: £19.50
State withholding: £8.78
Net pay: £69.22
```

````python
name = input("Employee name: ")
hours = float(input("Hours worked: "))
rate = float(input("Hourly rate: "))
federal_tax = float(input("Federal tax rate (e.g. 0.20 for 20%): "))
state_tax = float(input("State tax rate (e.g. 0.09 for 9%): "))

gross_pay = hours * rate
federal_withholding = gross_pay * federal_tax
state_withholding = gross_pay * state_tax
net_pay = gross_pay - federal_withholding - state_withholding

print(f"\nEmployee: {name}")
print(f"Gross pay: £{gross_pay:.2f}")
print(f"Federal withholding: £{federal_withholding:.2f}")
print(f"State withholding: £{state_withholding:.2f}")
print(f"Net pay: £{net_pay:.2f}")
````

 **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
st = int(input("How many students? "))

total = 0
passed = 0
highest = None
lowest = None

for i in range(1,st+1):
    m = int(input(f"Mark {i}: "))
    
    total = total + m
    if m >= 50:
        passed = passed + 1
        
    if highest is None or m > highest:
        highest = m
        
    if lowest is None or m < lowest:
        lowest = m

avg = total / st
failed = st - passed

print('Average:', avg)
print('Highest:', highest)
print('Lowest:', lowest)
print('Passed:', passed)
print('Failed:', failed)
````

## Student Marks Analysis 
In order to solve this problem, I used: input() for user input; type casting with int(); a list to store marks with append(); a for loop to process items; Boolean expressions like (m >= 50); comparison operators (>=, >, <, ==); control flow with if/elif/else; list functions max(), min(), len(); accumulation with += to sum marks and count passes; and formatted output with print() and f-strings.

---

Counter of students with marks ≥ 50.

**`lowest = None`**

### Loop Over Students

**`for i in range(1, st + 1):`**  
Iterates from 1 up to and including `st` to read each student’s mark once.

**`m = int(input(f"Mark {i}: "))`**  
Asks for mark *i* and converts it to an integer.

**`total = total + m`**  
Adds the current mark to the running total (for average later).

**`if m >= 50:`**  
Checks pass condition (mark ≥ 50).  
**`passed = passed + 1`**  
Increments the pass counter when the condition is true.

**`if highest is None or m > highest:`**  
On the first mark (`highest is None`) or whenever the new mark is larger, update the maximum.  
**`highest = m`**

**`if lowest is None or m < lowest:`**  
On the first mark (`lowest is None`) or whenever the new mark is smaller, update the minimum.  
**`lowest = m`**

---

### Post-Loop Calculations

**`avg = total / st`**  
Computes the average mark using the total from the loop and the student count.

**`failed = st - passed`**  
Students who failed are the remainder after subtracting the number who passed.

---

### Output

**`print('Average:', avg)`**  
Shows the average mark.

**Create a list of seven dice throws**

The throws should be randomly generated and add to the list
The list should be shuffle and print it out
Sort the list descending order and print it out
Count the occurrences and print them out
<br>


````python
import random

nums = []
for i in range(7):
    nums.append(random.randint(1, 6))

random.shuffle(nums)
nums.sort(reverse=True) 

print(nums)             

for i in range(1, 7):    
    print(i, nums.count(i)) 
````

## Week 4: Strings, Special Methods, Modules and Try-Except statment


Write a program that asks the user to enter two numbers and divides them. Use try-except to catch any ZeroDivisionError and ValueError that might occur (e.g., division by zero or invalid input).

````python
try:
    n1 = float(input("enter a numerator: "))
    n2 = float(input("enter a denominator: "))
    result = n1 / n2
except ValueError:
    print("Invalid input: please enter numbers.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
else:
    print("Result:", result)
````

Create a function (s) counting all lower case, upper case, digits, and special symbols from a given string and print them out.
   
```python

   example_password = "P@#yn26at^&i5ve"
```

````python
def c(s):
    lower =0
    upper =0
    digits =0
    special = 0
    for ch in s:
        if ch.islower():   lower  += 1
        elif ch.isupper(): upper  += 1
        elif ch.isdigit(): digits += 1
        else:              special += 1
    print("lower:", lower)
    print("upper:", upper)
    print("digits:", digits)
    print("special:", special)

example_password = "P@#yn26at^&i5ve"
c(example_password)
````

## Count Lowercase, Uppercase, Digits, and Specials 
In order to solve this problem, I used: a function with a parameter (c(s)); a for-loop to iterate over characters; control flow with if/elif/else; string methods islower(), isupper(), and isdigit(); counters with augmented assignment (+= 1); an else branch to count special symbols; and print() to display the counts for lower, upper, digits, and special.

---

#### `def c(s):`
**Defines a function** named `c` that takes a single string argument `s` (the text to analyze).

---

#### `lower = 0`
**Counter for lowercase letters**, starts at 0.

#### `upper = 0`
**Counter for uppercase letters**, starts at 0.

#### `digits = 0`
**Counter for numeric digits (0–9)**, starts at 0.

#### `special = 0`
**Counter for special (non-alphanumeric) characters**, starts at 0.

---

#### `for ch in s:`
**Loops through each character** `ch` in the input string `s`.

---

#### `if ch.islower():    lower += 1`
If the character `ch` is a **lowercase letter** (`a–z`), **increase** the `lower` counter by 1.

---

#### `elif ch.isupper():  upper += 1`
Else, if `ch` is an **uppercase letter** (`A–Z`), **increase** the `upper` counter by 1.

---

#### `elif ch.isdigit():  digits += 1`
Else, if `ch` is a **digit** (`0–9`), **increase** the `digits` counter by 1.

---

#### `else:               special += 1`
Otherwise, the character is a **special symbol** (space or punctuation, e.g., `@ # ^ & !`), so **increase** the `special` counter by 1.

---

#### `print("lower:", lower)`
**Prints the total number of lowercase letters** found.

#### `print("upper:", upper)`
**Prints the total number of uppercase letters** found.

#### `print("digits:", digits)`
**Prints the total number of digits** found.

#### `print("special:", special)`
**Prints the total number of special characters** found.

###  Adding Line Numbers to Python Zen

Create a Python program that reads the content from the `python_zen.txt` file and adds line numbers to each line. Each line in the output file should start with a line number, followed by a dot and a space, and then the original line from the input file. Skip numbering the first line as it is the title. Save the result into a new file called `numbered_zen.txt`.

**Steps:**
1. Open the `python_zen.txt` file for reading.
2. Skip the first line (which is the title) when numbering.
3. For each subsequent line, prepend the line number followed by a dot and a space.
4. Write the modified content to a new file called `numbered_zen.txt`.

<hr>

````python
# please download the python_zen.txt file in order for this code to work 
````

````python
with open('python_zen.txt', 'r') as f:
    lines = f.readlines()

title = lines[0]
numbered_lines = [title]

for i, line in enumerate(lines[2:],start=1):
    numbered_lines.append(f"{i}. {line}")

with open('numbered_zen.txt', 'w') as f:
    f.writelines(numbered_lines)
    
with open('numbered_zen.txt', 'r') as f:
    data=f.read()
    
print(data)
````
